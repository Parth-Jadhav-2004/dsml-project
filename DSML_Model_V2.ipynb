```python
def plot_all_graphs(df):
    num_cols = len(df.columns) - 1  # Exclude the first column (assuming it's not a numerical column)
    num_rows = (num_cols + 1) // 2  # Calculate the number of rows for subplots
    
    fig, axes = plt.subplots(num_rows, 4, figsize=(15, 5*num_rows))  # Create a figure with multiple subplots
    axes = axes.flatten()  # Flatten the 2D array of axes
    
    for i, col in enumerate(df.columns[1:]):  # Exclude the first column (assuming it's not a numerical column)
        # Plot different types of graphs based on the type of data
        if df[col].dtype == 'object':
            sns.countplot(x=col, data=df, ax=axes[i])
        else:
            sns.histplot(df[col], ax=axes[i], kde=True)  # Histogram for numerical data
        
        axes[i].set_title(col)  # Set title for the subplot
    
    # Remove any extra empty subplots
    for j in range(num_cols, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()

plot_all_graphs(df)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.50, random_state = 1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
list_met = []
list_accuracy = []
```

```python
#Model 1
#Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(C = 0.4, max_iter = 1000, solver = 'liblinear')
lr = classifier.fit(x_train, y_train)

#predition
y_pred = classifier.predict(x_test)

#accuracy
accuracy_LR = accuracy_score(y_test, y_pred)
```

```python
#Model 2
from sklearn.tree import DecisionTreeClassifier
classifier2 = DecisionTreeClassifier(random_state =14)
dt = classifier2.fit(x_train, y_train)

#prediction
y_pred2 = classifier2.predict(x_test)

#accuracy
accuracy_DT = accuracy_score(y_test, y_pred2)
```

```python
#model 3
from sklearn.ensemble import RandomForestClassifier
classifier3 = RandomForestClassifier(random_state = 14)
rfi = classifier3.fit(x_train, y_train)

#prediction
y_pred3 = classifier3.predict(x_test)

#accuracy
accuracy_RFI = accuracy_score(y_test, y_pred3)
```

```python
#model 4
classifier4 = RandomForestClassifier(criterion = 'entropy', random_state = 14)
rfe = classifier4.fit(x_train, y_train)

#prediction
y_pred4 = classifier4.predict(x_test)

#accuracy
accuracy_RFE = accuracy_score(y_test, y_pred4)
```

```python
#Model 5
from sklearn.svm import SVC

model_svm  = SVC()
svm = model_svm.fit(x_train, y_train)

#prediction
y_pred5 = model_svm.predict(x_test)

#accuracy
accuracy_SVC = accuracy_score(y_test, y_pred5)
```

```python
#MODEL 6
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
knn = model_knn.fit(x_train, y_train)

#prediction
pred_knn = model_knn.predict(x_test)
#accuracy
accuracy_knn = accuracy_score(y_test, pred_knn)
```

```python
#Model 7
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb = gnb.fit(x_train, y_train)

#prediction
pred_gnb = gnb.predict(x_test)

#accuracy
accuracy_GNB = accuracy_score(y_test, pred_gnb)
```

```python
#model 8
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
bnb = model.fit(x_train, y_train)

#prediction
pred_bnb = bnb.predict(x_test)

#accuracy
accuracy_BNB = accuracy_score(y_test, pred_bnb)
```

```python
#Combine all  the models by using Voting Classifier model
from sklearn.ensemble import VotingClassifier

evc = VotingClassifier(estimators = [('lr', lr), ('DT',dt), ('RFI', rfi), ('RFE', rfe), ('SVC', svm), ('KNN', knn), ('GNB', gnb), ('BNB', bnb)], voting = 'hard', flatten_transform = True)

model_evc = evc.fit(x_train, y_train)

#prediction
pred_evc = evc.predict(x_test)

#accuracy
accuracy_evc = accuracy_score(y_test, pred_evc)

list1 = ["Logistic Regression", "Decision Tree", "Random Forest Gini", "RandomForest Entropy", "Support Vector", "K nearest neighbour", "GaussianNB", "BermnoulliNB", "Ensemble Classifier"]

list2 = [accuracy_LR, accuracy_DT, accuracy_DT, accuracy_RFE, accuracy_SVC, accuracy_knn, accuracy_GNB,accuracy_BNB, accuracy_evc]

list3 = [classifier,classifier2,classifier3,classifier4, model_svm,model_knn, gnb, bnb, model, model_evc]

df_accuracy = pd.DataFrame({'Method used' :list1, "Accuracy": list2})
print(df_accuracy)
charts = sns.barplot(x = "Method used", y = "Accuracy", data = df_accuracy)
charts.set_xticklabels(charts.get_xticklabels(), rotation = 90)
```

```python
# roc and auc

def plot_roc(model , x_test , y_test):
    probablities = model.predict_proba(np.array(x_test))
    predictions =probablities
    fpr ,tpr , threshold =roc_curve(y_test , predictions[:,1])
    roc_auc = auc(fpr ,tpr)

    plt.title("Receiver operating charecteristic ")
    plt.plot(fpr , tpr,'b' , label='AUC =%0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0,1], [0,1], 'r--')
    plt.ylabel("True positive Rate")
    plt.xlabel("False positive Rate")
    plt.show()
```
